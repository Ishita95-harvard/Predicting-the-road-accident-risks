```python
import pandas as pd

# Load the training and test data
train_df = pd.read_csv('/content/train (6).csv')
test_df = pd.read_csv('/content/test (6).csv')
submission_df = pd.read_csv('/content/sample_submission (3).csv')

# Display the first 5 rows of the training data
print("First 5 rows of the training data:")
display(train_df.head())

# Display the first 5 rows of the test data
print("\nFirst 5 rows of the test data:")
display(test_df.head())

# Display the first 5 rows of the sample submission file
print("\nFirst 5 rows of the sample submission file:")
display(submission_df.head())
```

# Task
Analyze the provided dataset to predict accident risk and generate a submission file in the specified format.

## Data exploration and preprocessing

### Subtask:
Explore the datasets to understand the data types, distributions, and identify missing values. Handle any missing values and encode categorical features as needed.


**Reasoning**:
Display the data types and check for missing values for both train and test dataframes.



```python
print("Train DataFrame Info:")
train_df.info()

print("\nTest DataFrame Info:")
test_df.info()

print("\nMissing values in Train DataFrame:")
display(train_df.isnull().sum())

print("\nMissing values in Test DataFrame:")
display(test_df.isnull().sum())
```

**Reasoning**:
Display descriptive statistics for numerical columns and identify categorical columns in both dataframes.



```python
print("\nTrain DataFrame Descriptive Statistics:")
display(train_df.describe())

print("\nTest DataFrame Descriptive Statistics:")
display(test_df.describe())

categorical_cols_train = train_df.select_dtypes(include='object').columns
categorical_cols_test = test_df.select_dtypes(include='object').columns

print("\nCategorical columns in Train DataFrame:", list(categorical_cols_train))
print("Categorical columns in Test DataFrame:", list(categorical_cols_test))
```

**Reasoning**:
There are no missing values in the dataframes, so proceed with one-hot encoding the identified categorical features in both train and test dataframes and ensure they have the same columns.



```python
categorical_cols = ['road_type', 'lighting', 'weather', 'time_of_day']

train_df_encoded = pd.get_dummies(train_df, columns=categorical_cols, drop_first=True)
test_df_encoded = pd.get_dummies(test_df, columns=categorical_cols, drop_first=True)

# Align columns - crucial for consistent feature sets
train_cols = set(train_df_encoded.columns)
test_cols = set(test_df_encoded.columns)

missing_in_test = list(train_cols - test_cols)
for c in missing_in_test:
    test_df_encoded[c] = 0

missing_in_train = list(test_cols - train_cols)
for c in missing_in_train:
    train_df_encoded[c] = 0

# Ensure the order of columns is the same
test_df_encoded = test_df_encoded[train_df_encoded.columns]

print("\nTrain DataFrame after One-Hot Encoding and Alignment:")
display(train_df_encoded.head())

print("\nTest DataFrame after One-Hot Encoding and Alignment:")
display(test_df_encoded.head())
```

## Feature engineering

### Subtask:
Create new features that could potentially improve the model's performance.


**Reasoning**:
Create the new features based on the instructions and add them to both dataframes.



```python
# 1. Create an interaction feature between num_lanes and speed_limit
train_df_encoded['lanes_speed_interaction'] = train_df_encoded['num_lanes'] * train_df_encoded['speed_limit']
test_df_encoded['lanes_speed_interaction'] = test_df_encoded['num_lanes'] * test_df_encoded['speed_limit']

# 2. Create a polynomial feature by squaring the curvature column
train_df_encoded['curvature_squared'] = train_df_encoded['curvature']**2
test_df_encoded['curvature_squared'] = test_df_encoded['curvature']**2

# 3. Create a feature that represents the risk associated with num_reported_accidents by taking the square root
train_df_encoded['reported_accidents_risk'] = train_df_encoded['num_reported_accidents']**0.5
test_df_encoded['reported_accidents_risk'] = test_df_encoded['num_reported_accidents']**0.5

# Display the first few rows of the dataframes with the new features
print("Train DataFrame with new features:")
display(train_df_encoded.head())

print("\nTest DataFrame with new features:")
display(test_df_encoded.head())
```

## Model selection and training

### Subtask:
Choose a suitable model for the regression task and train it on the training data.


**Reasoning**:
Select and train a RandomForestRegressor model on the training data with the target variable separated.



```python
from sklearn.ensemble import RandomForestRegressor

# Separate features (X) and target variable (y)
X_train = train_df_encoded.drop(['id', 'accident_risk'], axis=1)
y_train = train_df_encoded['accident_risk']

# Initialize and train the RandomForestRegressor model
model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
model.fit(X_train, y_train)

print("Model training complete.")
```

## Prediction

### Subtask:
Make predictions on the test data using the trained model.


**Reasoning**:
Select the features from the test data that match the training data features and make predictions using the trained model.



```python
# Select the features for prediction from the test data, ensuring the column order matches X_train
X_test = test_df_encoded[X_train.columns]

# Make predictions on the test data
predictions = model.predict(X_test)

print("Predictions on test data complete.")
```

## Submission file generation

### Subtask:
Create a submission file in the specified format.


**Reasoning**:
Create the submission DataFrame, assign the predictions, and save it to a CSV file.



```python
# 1. Create a new DataFrame for the submission file using the submission_df as a template.
submission_output_df = submission_df.copy()

# 2. Assign the predictions made in the previous step to the 'accident_risk' column of the new submission DataFrame.
submission_output_df['accident_risk'] = predictions

# 3. Save the submission DataFrame to a CSV file named 'submission.csv'. Ensure that the index is not included in the output file.
submission_output_df.to_csv('submission.csv', index=False)

print("Submission file created successfully!")
```
